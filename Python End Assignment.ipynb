## PYTHON  ASSIGNMENT

```python
# Import Necessary Libraries

import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# Load and Preview Dataset

data = pd.read_excel ('myexcel.xlsx')
data.head (10)
```

## Preprocessing

```python
# Prepocessing Column Height

data['Height'] = np.random.randint(150, 181, size=len(data))
data.head(10)
```

```python
missing_data = data.isnull().sum()
missing_data
```

```python
# Handling missing values in  college column

data['College'].fillna(data['College'].mode()[0])
```

```python
# Handling missing values in  Salary column

data['Salary'].fillna(data['Salary'].median())
```

```python
# Handling missing values in  Age column

data['Age Group'].fillna(data['Age Group'].mode()[0])
```

```python
print(data.isnull().sum())
```

-------------------

## Analysis

```python
#  Task 1 - Distribution of employees across teams and percetange split

team_dist= data['Team'].value_counts()
team_prcnt = (team_dist / len(data)) * 100

team_sumry = pd.DataFrame({'Employee Count': team_dist, 'Percentage': team_prcnt})
team_sumry
```

```python
# Task 2 -Segregating Employees based on their position

position_dist = data['Position'].value_counts()

position_dist
```

```python
# Task 3 -Predominant Age group 

age_bins = [20, 30, 40, 50, 60, 70] 
# Adjust bins as needed

age_labels = ['20-30', '31-40', '41-50', '51-60', '61-70']

data['Age Group'] = pd.cut(data['Age'], bins=age_bins, labels=age_labels, right=False)
ag_grp_count = data['Age Group'].value_counts()
prednt_age = ag_grp_count.idxmax()

print("Predominant Age Group:", prednt_age)
print(ag_grp_count)

```

```python
# Task 4 -Team and position with highest salary expenditure.

team_salary_expenditure = data.groupby('Team')['Salary'].sum()

position_salary_expenditure = data.groupby('Position')['Salary'].sum()

hst_team = team_salary_expenditure.idxmax()
hst_team_exp = team_salary_expenditure.max()

hst_position = position_salary_expenditure.idxmax()
hst_position_exp = position_salary_expenditure.max()

print(f"Team with highest salary expenditure: {hst_team} with ${hst_team_exp}")
print(f"Position with highest salary expenditure: {hst_position} with ${hst_position_exp}")

```

```python
# Task 5 -Correlation between age and salary
# (Figure shown in Graphical Representation)

correlation = data['Age'].corr(data['Salary']) 

print("Correlation between Age and Salary:", correlation)
```

## Graphical Representation 

```python
#  Task 1 -Distribution of employees across teams and percetange split

plt.figure(figsize=(8, 5))
team_sumry['Employee Count'].plot(kind='bar', color='skyblue')
plt.title("Employee Distribution by Team")
plt.xlabel("Team")
plt.ylabel("Number of Employees")
plt.show()
```

```python
# Task 2 -Segregating Employees based on their position

plt.figure(figsize=(6, 4))
position_dist.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette("Set3"))  # autopct = %
plt.title("Employee Distribution by Position")
plt.ylabel('')  # Removes the 'y-axis' 
plt.show()

```

```python
# Task 3 -Predominant Age group 

plt.figure(figsize=(6, 4))
ag_grp_count.sort_index().plot(kind='bar', color='lightgreen')
plt.title("Employee Distribution by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.show()
```

```python
# Task 4 -Team and position with highest salary expenditure.

salary_team_position = data.groupby(['Team', 'Position'])['Salary'].sum().unstack()
salary_team_position.plot(kind='bar', stacked=True, figsize=(6, 4), colormap="viridis")
plt.title("Salary Expenditure by Team and Position")
plt.xlabel("Team")
plt.ylabel("Total Salary Expenditure")
plt.legend(title="Position")
plt.xticks(rotation=90) # NOT TO OVERLAP
plt.show()
```

```python
# Task 5 -Correlation between age and salary

plt.figure(figsize=(6, 5))
sns.scatterplot(data=data, x='Age', y='Salary')
plt.title("Age vs Salary Correlation")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```
